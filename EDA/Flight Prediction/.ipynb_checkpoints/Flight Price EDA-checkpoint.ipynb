```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
final_df = pd.read_excel("Data_train.xlsx")
final_df.head()
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df['Dept_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min'] = final_df['Dep_Time'].str.split(':').str[1]
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df['Dept_hour'] = final_df['Dept_hour'].astype(int)
final_df['Dept_min'] = final_df['Dept_min'].astype(int)
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stops':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1}) 
```

```python
final_df.head()
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Additional_Info'] = final_df['Additional_Info'].replace('No Info', 'No info')
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration_hour'] == '5m']
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype('int')
```

```python
final_df['Duration_hour'].isnull().sum()
```

```python
final_df['Duration_min'] = final_df['Duration']. str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_min'] = final_df['Duration_min'].fillna(0).astype('int')
```

```python
final_df['Duration_min'] = final_df['Duration_min'].astype('int')
```

```python
final_df['Total_duration'] = (final_df['Duration_hour']*60) + final_df['Duration_min']
```

```python
numerical_features = [features for features in final_df.columns if final_df[features].dtype!='O']
```

```python
numerical_features
```

```python
# Check for missing or NaN values
print(final_df[numerical_features].isnull().sum())
```

```python
#Histogram density plot

fig, axes = plt.subplots(len(numerical_features), 2, figsize=(14, len(numerical_features) * 4))

axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.histplot(data=final_df, x=feature, kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f'Histogram Density Plot for {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
